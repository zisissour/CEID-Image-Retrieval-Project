```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Load model results

with open('./results/pca_100', 'rb') as input:
    pca_100 = pickle.load(input)

with open('./results/pca_50', 'rb') as input:
    pca_50 = pickle.load(input)

with open('./results/pca_10', 'rb') as input:
    pca_10 = pickle.load(input)

with open('./results/ae_linear_100', 'rb') as input:
    ae_lin_100 = pickle.load(input)

with open('./results/ae_linear_50', 'rb') as input:
    ae_lin_50 = pickle.load(input)

with open('./results/ae_linear_10', 'rb') as input:
    ae_lin_10 = pickle.load(input)

with open('./results/ae_non_linear_100', 'rb') as input:
    ae_nonlin_100 = pickle.load(input)

with open('./results/ae_non_linear_50', 'rb') as input:
    ae_nonlin_50 = pickle.load(input)

with open('./results/ae_non_linear_10', 'rb') as input:
    ae_nonlin_10 = pickle.load(input)
```

```python
dimensions = [10, 50, 100]
pca_data = [pca_10, pca_50, pca_100]
ae_linear_data = [ae_lin_10, ae_lin_50, ae_lin_100]
ae_non_linear_data = [ae_nonlin_10, ae_nonlin_50, ae_nonlin_100]


# Plotting
bar_width = 0.2
index = np.arange(len(dimensions))

fig, ax = plt.subplots()

bar1 = ax.bar(index - bar_width, [data*10 for data in pca_data], bar_width, label='PCA')
bar2 = ax.bar(index, [data*10 for data in ae_linear_data], bar_width, label='AE Linear')
bar3 = ax.bar(index + bar_width, [data*10 for data in ae_non_linear_data], bar_width, label='AE Non-Linear')

ax.set_xlabel('Reduced Dimensions')
ax.set_ylabel('Succesful image retrieval (%)')
ax.set_title('Comparison of Models by Image Retrieval')
ax.set_xticks(index)
ax.set_xticklabels(dimensions)
ax.legend()

plt.savefig('./images/comparison.png')
plt.show()
```
